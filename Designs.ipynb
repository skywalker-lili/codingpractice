```python
# LeetCode
# 208. Implement Trie (Prefix Tree)
class TrieNode(object):
    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.children = {}
        self.word_end = False
        self.count = 0
        

class Trie(object):

    def __init__(self):
        self.root = TrieNode()
        self.root.word_end = True # to ensure that when search for empty string, all words comes out

    
    # Helper function for insert
    def _recur_insert(self, node, word):
        node.count += 1
        if len(word) == 0:
            node.word_end = True
        else:
            head = word[0]
            rest = word[1:]
            
            if head in node.children:
                self._recur_insert(node.children[head], rest)
            else:
                node.children[head] = TrieNode()
                self._recur_insert(node.children[head], rest)
    
    # Main function for insert
    def insert(self, word):
        """
        Inserts a word into the trie.
        :type word: str
        :rtype: void
        """
        self._recur_insert(self.root, word)
            
        
    # Helper function for search
    def _recur_search(self, node, word):
        #print "search for: {}".format(word)
        
        if len(word) == 0:
            if node.word_end:
                return True
            else:
                return False
        
        else:
            head = word[0]
            rest = word[1:]
            if head in node.children:
                return self._recur_search(node.children[head], rest)
            else:
                return False
    
    # Main search function
    def search(self, word):
        """
        Returns if the word is in the trie.
        :type word: str
        :rtype: bool
        """
        return self._recur_search(self.root, word)
        
    # Helper function
    def _recur_startsWith(self, node, prefix):
        
        if len(prefix) == 0:
            return True
        
        else:
            head = prefix[0]
            rest = prefix[1:]
            if head not in node.children:
                return False
            else:
                return self._recur_startsWith(node.children[head], rest)
    
    # Main function for 
    def startsWith(self, prefix):
        """
        Returns if there is any word in the trie
        that starts with the given prefix.
        :type prefix: str
        :rtype: bool
        """
        return self._recur_startsWith(self.root, prefix)
    
    
    def _words_counts(self, node, prefix, sofar, res):
        """
        Returns all words starting with this prefix and their counts
        """
        # if there is no prefix, add words whenever there is a word_end mark
        if len(prefix) == 0:
            if node.word_end:
                res.append((sofar, node.count))
            # For all subsequent child, add the words
            for head,child in node.children.items():
                self._words_counts(child, "", sofar+head, res)
        
        # if there is still prefix, keep searching
        else:
            head = prefix[0]
            rest = prefix[1:]
            if head in node.children:
                self._words_counts(node.children[head], rest, sofar+head, res)
        
    def autocomplete(self, prefix, k=5):
        """
        Returns the top k frequent words with that prefix
        """
        res = []
        self._words_counts(self.root, prefix, "", res)
        res.sort(reverse = True, key = lambda x: x[1])
        top_k = [res[i][0] for i in range(min(len(res), k))]
        
        return top_k

# Your Trie object will be instantiated and called as such:
# trie = Trie()
# trie.insert("somestring")
# trie.search("key")
```

```python
# Rate limiter

# Version 1.x: use a hit counter (LeetCode 362)
# Use an array to keep track of each second's hit and also the time of each hits
# Use mod to wrap around. 
# To get a hit count, loop over the whole hit array, sum an item if its hit time is later than the earlier hit that counts

class Rate_limiter1():
    def __init__(self, limit, per):
        self.hits_time = [None]*limit
        self.hit_i = 0
        self.limit = limit
        self.per = per
    
    def save_orDiscard(self, msg, hit_time):
        hit_time_limitAgo = self.hits_time[self.hit_i]
        if  hit_time_limitAgo == None or hit_time_limitAgo < hit_time - per:
            self.hits_time[self.hit_i] = hit_time
            self.hit_i = (self.hit_i + 1) % limit
            return "Save msg - {}".format(msg) # save_message()
        else:
            return "Ignore msg - {}".format(msg) # discard_message()
                

# Version 2: use token buckets
import time
class Rate_limiter2():
    def __init__(self, limit, per):
        self.limit = 1.0*limit; # unit: messages
        self.per  = 1.0*per; # unit: seconds
        self.allowance = 1.0*limit; # unit: messages
        self.last_check = 0.0; # floating-point, e.g. usec accuracy. Unit: seconds

    def save_orDiscard(self, msg, timestamp):
        # caculate the time from previous check
        time_passed = timestamp - self.last_check;
        self.last_check = timestamp; # update the last check
        
        # Check the allowance
        self.allowance += time_passed * (self.limit / self.per) #increase the allowance
        if (self.allowance > self.limit):
            self.allowance = self.limit; # maximum allowance
        
        # See if there are enough allowance for the new msg
        if (self.allowance < 1.0):
            return "Ignore this msg - {}".format(msg) # discard_message()
        else:
            self.allowance -= 1.0
            return "Save this msg - {}".format(msg) # save_message()
```

```python
# Hashtable
class MapEntry:
        """
        Inner class to fill in as the entry in hashtable
        """
        def __init__(self, key=None, val=None, is_occupied = False):
            self.key = key
            self.val = val
            self.is_occupied = is_occupied
        
        def __str__(self):
            return str((self.key, self.val, self.is_occupied))
        def __repr__(self):
            return self.__str__()

class Map:
    def __str__(self):
        return str("Size:{}, Occupied:{}, Array:{}".format(self.size, self.occupied, self.array))
    def __repr__(self):
        return self.__str__()
    
    def __init__(self, size = 5, load_factor = 0.5):
        self.array = [MapEntry() for _ in range(size)]
        self.size = size
        self.occupied = 0
        self.load_factor = load_factor
    
    def _hash(self, key):
        """
        Return the hash code for key
        """
        return (key % self.size)
    
    def add(self, key, val):
        """
        Add val into hashtable using key
        """
        hashcode = self._hash(key)
        prob = hashcode # start probing at the hashcode
        is_update = False
        
        # Find the first non-occupied postion or a position with same key using linear probing
        while self.array[prob].is_occupied:
            if self.array[prob].key != key:
                prob = (prob+1) % self.size
            else:
                is_update = True
                break
        
        # Insert val into found position
        entry = self.array[prob]
        entry.key, entry.val = key, val
        entry.is_occupied = True
        if not is_update: # only increase occupied when it's not an update
            self.occupied += 1
        
        # Check if the load factor has reached, if yes, enlarge the array
        # then copy the current value into the new array and use new array as the 
        if self.occupied*1.0/self.size > self.load_factor:
            self._enlarge()
    
    def _enlarge(self):
        """
        Trible the size of current arrayto reduce load factor
        """
        # Update the parameters, but remeber to keep track of the old array before finish
        temp = self.array
        new_array = [MapEntry() for _ in range(self.size*2)]
        self.array = new_array
        self.size = self.size * 2
        self.occupied = 0
        # Copy existing entries into new  array
        for i in range(len(temp)):
            entry = temp[i]
            if entry.is_occupied == True:
                self.add(entry.key, entry.val) # re-add to keep the key-value mapping working in new size!
        # Done!
    
    def get(self, key):
        """
        Get the value corresponding to
        """
        hashcode = self._hash(key)
        prob = hashcode
        while self.array[prob].is_occupied:
            # Check if this entry has the same key as what we are looking for
            if self.array[prob].key == key:
                return self.array[prob].val
            else:
                prob = (prob+1) % self.size
        
        return None # prob reaches an empty entry, the key doesn't exist
            
    def delete(self, key):
        """
        Delete the item corresponding to key. Do nothing if the item doesn't exist
        """
        hashcode = self._hash(key)
        prob = hashcode
        while self.array[prob].is_occupied:
            # Set the entry to be non-occupied
            if self.array[prob].key == key:
                self.array[prob].is_occupied = False
                self.occupied -= 1
                break
            else:
                prob = (prob+1) % self.size
```
