```python
# binary search
# array need to be sorted
def binary_search(array, key):
    if len(array) == 0: return False, -1
    start = 0
    end = len(array)
    while start < end:
        mid = (start + end)/2
        if array[mid] == key: 
            return True, mid
        elif array[mid] < key: 
            start = mid + 1
        else: 
            end = mid
    #if start == key: return True, start
    return False, -1

# find the first item that equals key
def binary_search_first(array, key):
    if len(array) == 0: return False, -1
    start = 0
    end = len(array)
    while start < end:
        mid = (start + end)/2
        if array[mid] == key:
            if mid == 0: return True, mid
            elif array[mid-1] < array[mid]: 
                return True, mid
            else:
                end = mid
        elif array[mid] < key: start = mid + 1
        else: end = mid
    return False, -1

# find the last item that equals key
def binary_search_last(array, key):
    if len(array) == 0: return False, -1
    start = 0
    end = len(array)
    while start < end:
        mid = (start + end)/2
        if array[mid] == key:
            if mid == len(array)-1: return True, mid
            elif array[mid+1] > array[mid]: 
                return True, mid
            else:
                start = mid + 1
        elif array[mid] < key: start = mid + 1
        else: end = mid
    return False, -1
```

```python
# Insertion sort

def swap(array, i, j):
    temp = array[i]
    array[i] = array[j]
    array[j] = temp

def insertion_sort(array):
    if len(array) == 0: return
    
    for i in range(1,len(array)):
        j = i
        while array[j] < array[j-1] and j >= 1:
            swap(array, j-1, j)
            j = j-1
    return array
```

```python
# Selection sort

def find_min(array, s, t):
    min_index = s
    for i in range(s, t):
        if array[i] < array[min_index]:
            min_index = i
    return min_index

def swap(array, i, j):
    temp = array[i]
    array[i] = array[j]
    array[j] = temp

def selection_sort(array):
    if len(array) == 0: return
    
    for i in range(len(array)):
        min_index = find_min(array, i, len(array))
        swap(array, i, min_index)
    
    return array
```

```python
# Bubble sort

def bubble_sort(lst, h, k):
    if h == k: return
    for n in range(k,n,-1):
        for i in range(h,n):
            if lst[i] > lst[i+1]:
                swap(lst, i, i+1)
   
```

```python
# QuickSort
# main procedure
def QS(lst, h, k):
    if h == k: return
    pivot = partition(lst, h, k)
    QS(lst, h, pivot-1)
    QS(lst, pivot+1, k)

# partition a list
def partition(lst, h, k):
    start = h
    end = k
    
    # 默认以lst的h为pivot
    # 不停地把比pivot小或者等于地元素放到pivot地左边
    # 把比pivot大的元素放到右边
    while start < end:
        if lst[start+1] <= lst[start]:
            swap(lst, start+1, start)
            start += 1
        else:
            swap(lst, start+1, end)
            end -= 1
    return start

# swap two elements in a list
def swap(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
```

```python
# MergeSort
# main procedure to merge-sort  a list from element h to elsemt k (including)
def MS(lst, h, k):
    if h = k: return
    mid = (h+k)/2
    MS(lst, h, mid)
    MS(lst, mid+1, k)
    merge(lst, h, mid, k)

# merge two sorted part of a list
def merge(lst, h, t, k):
    temp = lst[h: t+1] # a temporary
    i, j = 0, t+1 # two pointers
    for m in range(h, k+1):
        if temp[i] <= lst[j]:
            lst[m] = temp[i]
            i += 1
        else:
            lst[m] = lst[j]
            j += 1
```

```python
# pre-order traverse a tree
def DFS(root):
    visited = list()
    if root == None:
        return
    else:
        visited.append(root)
        DFS(root.left)
        DFS(root.right)
    return visited()
    
# Using a while loop
# first visit the given node; then instead of visit the left, reset the node as it's left child
# so that the while loop can keep visiting the new node which is the left child of the upper level node
# use a stack to stack up right children as visiting a node so that only after all left nodes visited, we can start visit right nodes
# stacking up the right nodes should happen before resetting node so that the right nodes will be visited according to how left they are
def preorderTraverse(root):
    s = list()
    while len(s) > 0 or root != None:
        if root != None:
            visit(root)
            if root.right != None:
                s.append(root.right)
            root = root.left # this step is robust because when the left is None, root is set to be nodes in stack 
        else:
            root = s.pop()
```

```python
# in-order traverse a tree
def DFS(root):
    visited = list()
    if root == None:
        return
    else:
        DFS(root.left)
        visited.append(root)
        DFS(root.right)
    return visited

# For a given tree, visit the left part then root then right part
# actual visit happens when a node has no child at all
def inorderTraverse(root):
    s = list()
    visited = list()
    while len(s) > 0 or root != None:
        if root != None:
            s.append(root)
            root = root.left # while root has left, search that left before visiting
        else: # the new root is None, the previous root must be the left most
            visited.append(s.pop()) # visit that left most root
            root = root.right # then try to visit right after visit the root
    return visited
```

```python
# post-order traverse a tree
def DFS(root):
    visited = list()
    if root == None:
        return
    else:
        DFS(root.left)
        DFS(root.right)
        visit(root)
    return visited
        
# For a given tree, visit the left part then left then right part
# For each tree, the algorithm should try to search for the left child then right child; until there's no 
# there should be an machanism to trace back to the parent node and try visit the parent's other child first
#   tracing back can be done using stack
#   trying algorithm realized by put the visit after if-conditions
#      visit a node only when it's left node and right node are visited
#        left node visited is ensured because it only goes back from 
def postorderTraverse(root):
    s = list() # list of not visited nodes
    visited = list() # list of visited nodes
    while len(s) > 0 or root != None:
        if root != None:
            s.append(root)
            root = root.left
        else: # pointing to None
            visit.append(s.pop()) # the privous root is the left most node, visit it
            root = s[-1] # goes up to parent node but No pop because the parent aren't visited yet
            if root.right != None and root.right not in visited:
                s.append(root.right)
            else:
                visit.append(s.pop()) # this parent doesn't have right child, it's time to visit
```

```python
# level-order traverse a tree
def BFS(root):
    q = Queue()
    q.put(root)
    while len(q) > 0:
        node = q.get()
        visit(node)
        for child in node.children:
            if child != None:
                q.put(child)
    return q
```

```python
# Binary search

def BS(lst, s, e, num):
    if len(lst[s:e]) == 0: return False   
    
    while s < e: # s-e = number of items to be searched!
        m = (s+e)/2
        if lst[m] < num:
            s = m+1
        elif lst[m] > num:
            e = m-1
        else: # lst[m] = num
            return m
    return False


def BS_first(lst, s, e, num):
    if len(lst[s:e]) == 0: return False
    while s < e:
        m = (s+e)/2
        if lst[m] < num:
            s = m+1
        elif lst[m] > num:
            e = m-1
        elif lst[m] == lst[m-1]:
            e = m-1
        else:
            return m
    return False
```

```python
# Backtracking
# LeetCode
# 332. Reconstruct Itinerary
# --------------------------

class Solution(object):
    def findItinerary(self, tickets):
        """
        :type tickets: List[List[str]]
        :rtype: List[str]
        """
        n = len(tickets)
        path = ['JFK'] # initial path contains 'JFK'
        
        # make a dictionary based on lists of tickets, for each starting airport, the list of destinations is sorted
        # like: {"JFK": ["ABC", "ACC"...""KFJ"]}
        def make_dic(tickets):
            dic = {}
            for ticket in tickets:
                if ticket[0] not in dic.keys():
                    dic[ticket[0]] = [ticket[1]]
                else:
                    dic[ticket[0]].append(ticket[1])
            
            for val in dic.values():
                val.sort()
                
            return dic
        
        dic = make_dic(tickets)
        
        def dfs(path):
            # if path is equal to (n+1): path discovered, return True
            if len(path) == n+1: return True
            
            # while path not discovered yet
            cur_airport = path[-1]
            num_potential_dests = len(dic[cur_airport]) # index of potential destinations can choose
            
            for i in range(num_potential_dests): # try all possible next destinations 
                ticket_to_use = dic[cur_airport].pop(0) # use the next ticket from the sorted list of 
                path.append(ticket_to_use) # update the path
                path_found = dfs(path)
                if path_found: 
                    return True # found a path that is alphabetically sorted! return True
                else:
                    # path doesn't find, need to "return" the ticket and go back to the previous airport
                    dic[cur_airport].append(ticket_to_use)
                    path.pop()
            
            # after for loop,if path not found, that means given the current path
            # there is no solution 
            return False
        
        dfs(path)
        
        return path
```

```python
# Dynamic programming
# Subset sum

def subset_sum(ints, s):
    
    # Base case, when s = 0, of course can find the sum
    if s == 0: return True
    
    # Build a matrix holding the possible answers
    solution = [[None]*(s+1) for dum_i in range(len(ints)+1)]
    
    # Fill out the base cases:
    # 1. solution[:][0] are True because we can get return empty set
    n = len(ints)
    for i in range(n+1):
        solution[i][0] = True
    # 2. solution[0][1:] are False because we can't produce sum out of nothing
    for j in range(1:s+1):
        solution[0][j] = False
    
    # Produce the solution matrix
    for i in range(1:n+1):
        for j in range(1:s+1):
            if solution[i-1][j]: # if the above cell is true, this is also true
                solution[i][j] = True
            elif j - ints[i-1] >= 0: # above cell is false, but we still have hope see if including the current 
                solution[i][j] = False or solution[i-1][j - ints[i-1]]
            else:
                solution[i][j] = False
    
    return solution[n][s]
```

```python
# Dynamic programming
# Maximum Subarray 

def max_subarray(ints):
    
    # initialize the array holding maximum sum of arrays ending at index i, i = 1:len(ints)
    solution= [0]*(len(ints)+1)
    
    for i in range(len(ints)):
        # Either including the current int at ints[i] into the sum array OR
        # just set it to zero, indicating we don't have a array ending at this position
        solution[i+1] = max(solution[i]+ints[i], 0)
    
    return max(solution)

#print max_subarray([1,2,-3])
```

```python
# Diameter Of a Binary Tree
# http://algorithms.tutorialhorizon.com/diameter-of-a-binary-tree/

def find_height(root):
    if root == None: return 0
    else:
        return max(1+find_height(root.left), 1+find_height(root.right))

def find_diameter(root):
    
    if root == None: return 0
    
    longest_path_btw = 1 + find_height(root.left) + find_height(root.right)
    longest_path_left = find_diameter(root.left)
    longest_path_right = find_diameter(root.right)
    
    return max(longest_path_btw, longest_path_left, longest_path_right)
```

```python
# In-order traverse of binary search tree iteration
# http://algorithms.tutorialhorizon.com/inorder-traversal-non-recursive-approach/

def in_order(root):
    stack = []
    while True:
        # Go to left most
        while root != None:
            stack.append(root)
            root = root.left
        
        # If stack is empty, return
        if len(stack) == 0: return
        
        # Pop out the left most one
        leftmost = stack.pop()
        do_something(leftmost)
        
        # No need to care about the root of leftmost
        # because this root will be pop up as long as leftmost doesn't have child
        # Only thing wo worry about is if leftmost has a child:
        root = leftmost.right # if leftmost one has right child, do something over there
```

```python
# Pre-order traverse of binary search tree iteration
# http://algorithms.tutorialhorizon.com/binary-tree-preorder-traversal-non-recursive-approach/

def pre_order(root):
    stack = []
    while True:
        # Go to left most
        while root != None:
            do_something(root)
            stack.append(root)
            root = root.left
        
        # If stack is empty, return
        if len(stack) == 0: return
        
        # Pop out the left most
        # since leftmost must be a root of some tree, it' already done
        # Only thing to worry about is if leftmost has a child
        leftmost = stack.pop()
        root = leftmost.right
```

```python
# Post-order traverse of binary search tree iteration
# http://algorithms.tutorialhorizon.com/binary-tree-postorder-traversal-non-recursive-approach/

# Basic idea is if a root is more "rooty" it should be executed first

def post_order(root):
    stack1 = []
    stack2 = []
    stack1.append(root)
    while len(stack1) > 0:
        node = stack1.pop()
        stack2.append(node)
    if node.left != None: stack1.append(node.left)
    if node.right != None: stack1.append(node.right)
```
