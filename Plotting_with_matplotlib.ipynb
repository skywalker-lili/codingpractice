## Expansions and Contractions

Multiple series can be plotted (or filled) using one of:

* `plot(x1,y1,x2,y2)`
* `plot(x1,y1,f1,x2,y2,f2)`
* `plot(y1,f1,y2,f2)`

where the `f`_i_ are format strings. See `plt.plot?` for details on the format strings, which can take colors, markers and line styles.  This example plots the expansions separately from the contractions as fill, where the expansions are green and the contractions are read.

```python
plt.figure()
expansion = real_gdp_growth.copy()
contraction = real_gdp_growth.copy()
expansion[expansion < 0] = 0
contraction[contraction > 0] = 0
expansion = np.concatenate(([0], expansion, [0]))
contraction = np.concatenate(([0], contraction, [0]))
plt.fill(expansion, 'g', contraction, 'r')
plt.autoscale(tight='x')
plt.draw()
```

Format strings can also be web colors, which provide precise color control. In this example, the keyword argument `linewidth` is set to 0 to prevent the line from being drawn.

```python
plt.figure()
plt.fill(expansion, '#00AA00', contraction, '#880000', linewidth=0)
plt.autoscale(tight='x')
plt.draw()
```

```python
# TODO: Good candidate to beautify, legend, title, ticks, etc.
# TODO: Add as problem to tasks
plt.figure()
years = np.arange(2009,2013)
counts = np.zeros((4,4))
for j,year in enumerate(years):
    recent_growth = world_bank_data.xs(str(year),level=1).growth
    recent_growth = recent_growth.dropna()
    for i,b in enumerate(breaks):
        counts[i,j] = np.sum((recent_growth > b[0]) & (recent_growth < b[1]))
```

There is no standard multiple bar chart command, so producing once with $n$ requires calling bar $n$ times.  The most important keyword argument is `width` which should be $1/(n+1)$.  The bar charts use the same widths and use the width to stagger the bars.

## 3-d plotting

While matplotlib is not the most complete 3-d plotting package for Python, it is almost always more than adequate since most 3-d plots are not needed (and often do more to obscure than convey useful information).

This example computes the bivariate Normal pdf for with correlation of 50% and unit variances across a domain of $[-3,3] \times [-3,3]$.

```python
def bivariate_normal_pdf(x,y,mu,sigma):
    sigma_inv = np.mat(np.linalg.inv(sigma))
    z = np.matrix([x,y]) - np.mat(mu)
    ll = 1./(2*np.pi) * 1./np.sqrt(np.linalg.det(sigma))
    ll *= np.exp(-0.5*z*sigma_inv*z.T)
    return  ll

mu = np.zeros(2)
sigma = np.ones((2,2)) * 0.5 + np.eye(2) * 0.5
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
pdf = np.zeros((100,100))
for i in xrange(100):
    for j in xrange(100):
        pdf[i,j] = bivariate_normal_pdf(x[i],y[j],mu,sigma)
```

## Multiple Axes on the same figure

Some graphs benefit form having a left and right y-axis to show data with different scale. `twinx` makes this relatively simple by copying the x-values to a new axis.  The basic strategy is to create a plot, and then get its exes from the _handle_ that is returned from `plot`.  This handle can then be used to get the line's axes using `get_axes`, and finally `twinx` is used to twin the axes.

```python
plt.figure()
output_gap = np.log(gdp_data.real_gdppot / gdp_data.real_gdp)
output_gap.name = 'output_gap'
common = gdp_data.copy()
common['output_gap'] = output_gap
common = common.dropna()
lines = plt.plot(common.index, common.real_gdp,'#990000',
              common.index, common.real_gdppot,'#444444',
              label=('Real GDP','Real Potential GDP'))
ax = lines[0].get_axes()
plt.ylabel('Real GDP')
ax2 = plt.twinx(ax)
ax2.plot(common.index,common.output_gap,label='Output Gap')
plt.ylabel('Output Gap')
plt.show()
```

## Heat Maps

Heat maps a simple method to show correlations or other dependence measures.  This example visualized the correlation among the stocks in the CRSP database example.  This first block computes correlations from those with no missing values.

```python
returns =  crsp_returns.dropna(axis=1)
R = returns.corr()
```

This next block computes the first principle component and the regression coefficient of the stock on this PC.

```python
Rval = R.values
lam,X = np.linalg.eigh(Rval)
f= X.T.dot(returns.values.T)
main_factor = f[-1,:]
beta = np.linalg.pinv(main_factor[:,None]).dot(returns)
```

Finally, the $R^2$ of the regression of the stock on the first PC is computed, and the panel is ordered from highest to lowest.  This is a simple method to get stocks that are more similar to the start of the list.

```python
errors = returns - main_factor[:,None].dot(beta)
R2 = 1 - np.sum(errors ** 2.0,0) / np.sum(returns ** 2.0)
order = np.argsort(R2)
R = returns[returns.columns[order.values[::-1]]].corr()
```

Finally, `imshow` can be used to visualize the correlation.  It takes data between 0 and 1, to the correlations are linearly shrunk from $[-1,1]$ to $[0,1]$.  Finally, a `colorbar` is added which is like a legend.

```python
fig, ax = plt.subplots()
# Rescale R to be between 0 (real corr = -1) and 1 (real corr = 1)
cax = ax.imshow((R+1.0)/2.0, interpolation='nearest', cmap=cm.hot_r)
ax.set_title('Correlation of 1500 stocks')

# Ticks have been rescale, so all between 0 and 1
cbar = fig.colorbar(cax, ticks=[0.3,0.5,.75 ,1])
# Tick labels match ticks
tl = cbar.ax.set_yticklabels(['-.4','0','0.5','1']) 
```

## Pie Charts and Further Customization

Pie charts are created with `pie` which, at a minimum, requires size data (normalized or not).  Other keyword arguments are `explode` which will push one or more slices from the center, `labels`, the list of slice labels, `colors`, a list of colors to use, `startangle` which will rotate the chart and `pctdistance` which controls the location of the number.

#### Handle Graphics
matplotlib uses handle graphics where every object - a line, a label or even the figure - has a handle that contains all of the information about this object.  This example shows one simple application of handle graphics to customize the color and size of the text labels.

```python
colors = seaborn.dark_palette("MediumPurple",4)

labels = ('Strong\nContraction','Contraction','Expansion','Strong\nExpansion')
sizes = counts[:,0]
explode = (.2,0,0,0) 


h = plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=270, pctdistance=0.8)

for i in xrange(4):
    h[2][i].set_color('#FFFFFF')
    h[2][i].set_fontsize(14)
    h[2][i].set_fontweight('bold')
    h[1][i].set_fontsize(14)

t = plt.title('2009 Cross-country Growth Rates')
t.set_fontsize(16)
plt.tight_layout()
plt.axis('equal')
plt.show()
```

## Fonts and $\LaTeX$

An alternative method to set font properties is to use a dictionary containing keys with specific names (more or less the same as CSS Level 1) and values.  Many of the functions that produce text labels accept a keyword argument `fontdict` which will take and process this dictionary.

```python
import numpy as np
import matplotlib.pyplot as plt

gray = '#666666'

font = {'family' : 'serif',
        'color'  : gray,
        'weight' : 'normal',
        'size'   : 16,
        }

h = plt.plot(gdp_growth.index, 100 * gdp_growth, '#210076')
plt.title('Annual Real GDP Growth, $100 \ln(RGDP_t/RGDP_{t-4})$', fontdict=font)
plt.xlabel('Date', fontdict=font)
plt.ylabel('Growth Rate', fontdict=font)
plt.autoscale()
plt.tight_layout()

ax = h[0].get_axes()
ticklabels = ax.get_xticklabels() + ax.get_yticklabels()
for label in ticklabels:
    label.set_family('serif')
    label.set_fontsize(12)
    label.set_color(gray)
    
plt.show()
```
